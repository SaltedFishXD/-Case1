```python
import math, re
from collections import Counter
import nltk
```

```python
def words(text): return re.findall(r'\w+', text.lower())
def bigrams(text): 
    lst0 = words(text)
    lst=[]
    for i in range(0,len(words(text))-1):
        lst.append(lst0[i]+' '+lst0[i+1])
    return lst
```
